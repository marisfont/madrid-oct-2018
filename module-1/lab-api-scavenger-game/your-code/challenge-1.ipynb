## Step 1: 
**Obtain all the forks of the ironhack-datalabs/madrid-oct-2018 through Github API.**

- I imported the json and requests library

```python
import json
import requests
```

- I created the variable response using the request.get to send an HTTP requests 
- I created the variable results to receive back a response in the form of JSON structured data

```python
response = requests.get('https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/forks')
results = response.json()
results
```

- I saved the results since we were having connection problems

```python
with open('challenge_1.json', 'w') as outfile:
   json.dump(results, outfile)
```

- I imported the pandas library

```python
import pandas as pd
```

- I created a data frame using pandas called data to analyze the data

```python
data = pd.DataFrame(results)
data
```
